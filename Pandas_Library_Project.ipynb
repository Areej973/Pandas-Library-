# Pandas Library Project
This project aims to enhance your understanding of the Pandas library concepts and your data-handling skills using it.


```python
import pandas as pd
import numpy as np
```

### Create a DataFrame with 15 rows and 3 columns using the pd.DataFrame function, and fill it with the following values.


```python
data = {'Level':[100, 100, 100, 200,np.nan, 200, 300, 300, np.nan],
        'Course Name':['Introduction to programming fundamentals','introduction to it','Introduuction to MIS ','Database management','Data structures','Systems analysis and design','Network securtity','Algorithms', np.nan],
        'Major':['IT','CS','MIS', np.nan, 'CS', 'MIS', np.nan , 'CS', np.nan]
        }

df = pd.DataFrame(data)
df
```

### Print all DataFrame properties, including describe.

```python
print('data properties:\n')
print(df.describe())
```

### Modify the data type of the Level column to be of type double.

```python
df['Level']=df['Level'].astype(float)
df.info()
```

### Print the first four rows of the DataFrame.

```python
print('The fist 4 rows:\n', df.head(4))
```

### Delete all rows that contain null values entirely.

```python
df=df.dropna(how='all')
```
